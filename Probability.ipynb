Frequency vs Probablity 
Before understanding Probability let us understand these terms:

Freqency: The frequnecy of a value is the number of times the observation occurs. Frequency is usally denoted by f

Relative Frequency: The ratio of the frequency to the total number of observation.It is denoted by rf rf=f/n

Cumulative Freqency: It gives the number of observations less than or equal to a specified value.It is usally denoted by cf

```python
#Let us create a random natural number from 1 to 100 and let us see their frequency
NN=np.random.randint(1,100,size=(100))
DataMatrix=pd.DataFrame({'RandomNaturalNumber':NN})
# let us create a frequency column 
DataMatrix["Frequency(no of times a value occured)"]=DataMatrix['RandomNaturalNumber'].value_counts()
# value_counts the unique values of the RandomNaturalNumber columns 
# we can clearly see that we can see NAN value which means we did not see any 0
DataMatrix.sort_values(by="Frequency(no of times a value occured)",ascending=False,inplace=True)
# Now we sorted our dataframe by frequency column in descending order
# all the NaN values suggest that we did not get any values (99,86,72,47) with Nan
# Now we will replace the NaN with zero
DataMatrix.fillna(0,inplace=True)


# Now let us create a Relative frequency of the dataframe
DataMatrix['RelativeFrequency(Frequency/totalNumberofObs)']=DataMatrix['Frequency(no of times a value occured)']/DataMatrix['RandomNaturalNumber'].count()
# Now let us Understand Relative frequency in percentage
DataMatrix['RelativeFrequencyinpercentage(RelativeFrequency *100)(%)']=DataMatrix['RelativeFrequency(Frequency/totalNumberofObs)']*100
DataMatrix
```

```python
No_of_Heads=[]
No_of_toss=[]
n=1
while n<501: # n represents the no of tosses
    Heads_per_toss=[]  # This one takes the values of 
    #print(m)     
    for i in range(0,n,1):
        #print(i)
        x=np.random.randint(0,2)
        #print(x)
        if x==1:
            Heads_per_toss.append(x)
        else:
            Heads_per_toss.append(x)
    #print(m)
        #print(Heads_per_toss)
    No_of_Heads.append(Heads_per_toss.count(1))
    No_of_toss.append(n)
    n=n+1
```

```python
# Now let us create dataframe of no.of tosses, no.of heads and probability of heads and percentage of probability
Tosses=pd.DataFrame({'No_of_toss':No_of_toss,'No_of_Heads':No_of_Heads})
# Create a column of probability of head and percentage of head
Tosses['Probability_of_Heads']=Tosses['No_of_Heads']/Tosses['No_of_Heads'].count()
Tosses['Percentage_of_Heads ']=Tosses['Probability_of_Heads']*100
Tosses
```

As we can clearly see and we increase our number of trails and the random variable starts converging(I Meaning getting close to 0.5 or 50 %) which says if you toss a coin more number of times your chance of getting a heads is 50 and tail is 50 and that is why a coin is called fair coin

```python
ax=sns.lineplot(data=Tosses,x='No_of_toss',y='Probability_of_Heads')
ax.set_ylim(0,1)
```
